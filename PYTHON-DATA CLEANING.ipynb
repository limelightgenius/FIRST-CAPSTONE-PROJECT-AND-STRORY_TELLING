```python
##SPLITTING COLUMNS
##REMOVING OR FILTERING OUT BAD OR CANCELLED ORDERS
## HANDLING OUTLIERS
##MERGING TABLES





                    FULL DAY 34 CAPSTONE PROJECT TEMPLATE 
import pandas as pd

# 1. Load
df = pd.read_csv("sales_data.csv")

# 2. Inspect
df.head(); df.info(); df.describe()

# 3. Missing values
df['PRICE_EACH'].fillna(df['PRICE_EACH'].median(), inplace=True)
df['QUANTITY_ORD'].fillna(df['QUANTITY_ORD'].median(), inplace=True)
df.dropna(subset=['SALES', 'ORD_DATE'], inplace=True)

# 4. Drop duplicates
df.drop_duplicates(inplace=True)

# 5. Fix data types
df['QUANTITY_ORD'] = df['QUANTITY_ORD'].astype(int)
df['PRICE_EACH'] = df['PRICE_EACH'].astype(float)
df['SALES'] = df['SALES'].astype(float)
df['ORD_DATE'] = pd.to_datetime(df['ORD_DATE'])

# 6. Trim + standardize text
df['PRODUCT_LINE'] = df['PRODUCT_LINE'].str.strip().str.title()
df['STATUS'] = df['STATUS'].str.strip().str.lower()
df['CUSTOMER_NAME'] = df['CUSTOMER_NAME'].str.strip().str.title()

# 7. Fix inconsistent
df['STATUS'].replace({
    'shiped':'shipped',
    'shippedd':'shipped',
    'canceled':'cancelled'
}, inplace=True)

# 8. Create new columns
df['Revenue'] = df['QUANTITY_ORD'] * df['PRICE_EACH']
df['Year'] = df['ORD_DATE'].dt.year
df['Month'] = df['ORD_DATE'].dt.month
df['Quarter'] = df['ORD_DATE'].dt.quarter

# 9. Split customer name
df[['FirstName','LastName']] = df['CUSTOMER_NAME'].str.split(' ', 1, expand=True)

# 10. Handle outliers (example)
df = df[df['QUANTITY_ORD'] < df['QUANTITY_ORD'].quantile(0.99)]
df = df[df['PRICE_EACH'] < df['PRICE_EACH'].quantile(0.99)]

# 11. Filter unwanted rows
df = df[df['STATUS'] != 'cancelled']
df = df[df['SALES'] > 0]

# 12. Remove irrelevant columns
df.drop(['ADDRESS_LINE2','PHONE'], axis=1, inplace=True)

# 13. Rename columns
df.rename(columns={
    'ORD_NUMBER': 'OrderNumber',
    'QUANTITY_ORD': 'Quantity',
    'PRICE_EACH': 'UnitPrice',
    'ORD_LINE_NUMBER': 'LineNumber',
    'ORD_DATE': 'OrderDate'
}, inplace=True)

# 14. Save cleaned dataset
df.to_csv("cleaned_sales_data.csv", index=False)

```
